```python
X = df1.drop('Criminal', axis='columns')
X.head()
```

```python
from sklearn.cluster import MiniBatchKMeans
```

```python
kmeans = MiniBatchKMeans(n_clusters=3, batch_size=3, max_iter=3).fit(X)
```

```python
len(kmeans.cluster_centers_)
```

```python
one = kmeans.cluster_centers_[0]
```

```python
two = kmeans.cluster_centers_[1]
```

```python
three =kmeans.cluster_centers_[2]
```

```python
one
```

```python
one_scores = {}

one_scores['personality_traits'] =  one[0] - one[1] + one[2]
one_scores['social_traits'] =  one[0] - one[1] + one[2]
one_scores['ethical_traits'] =  one[0] - one[1] + one[2]
```

```python
one_scores
```

```python
all_traits = {'one':one, 'two':two, 'three':three}

all_traits_scores = {}

for name, traits in all_traits.items():
    crime_trait = {}
    
    crime_trait['personality_traits'] =  traits[0] - traits[1] + traits[2]
    
    crime_trait['social_traits'] =  traits[0] - traits[1] + traits[2]
    
    crime_trait['ethical_traits'] =  traits[0] - traits[1] + traits[2]
    
    
    all_traits_scores[name] = crime_trait
    
    
```

```python
all_traits_scores
```

```python
personality = []
social = []
ethical = []

for traits, crime_trait in all_traits_scores.items():
    personality.append(crime_trait['personality_traits'])
    social.append(crime_trait['social_traits'])
    ethical.append(crime_trait['ethical_traits'])

```

```python
personality_normalized = (personality-min(personality))/(max(personality)-min(personality))

social_normalized = (social-min(social))/(max(social)-min(social))

ethical_normalized = (ethical-min(ethical))/(max(ethical)-min(ethical))

```

```python
personality_normalized
```

```python
counter = 0

normalized_all_traits = {}

for traits, crime_trait in all_traits_scores.items():
    normalized_trait = {}
    normalized_trait['personality_traits'] = personality_normalized[counter]
    
    normalized_trait['social_traits'] = social_normalized[counter]
    
    normalized_trait['ethical_traits'] = ethical_normalized[counter]
    
    normalized_all_traits[traits] = normalized_trait
    
    counter+=1
    
```

```python
normalized_all_traits
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(5,3))
plt.ylim(0,1)
plt.bar(list(normalized_all_traits['one'].keys()), normalized_all_traits['one'].values(), color='g')
plt.show()
```
